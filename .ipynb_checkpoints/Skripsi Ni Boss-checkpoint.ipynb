# Menghapus special karakter dan url

```python
def remove_text_special(text):
    # hapus tab, new line, ans back slice
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    # hapus non ASCII (emot, bahasa china dll)
    text = text.encode('ascii', 'replace').decode('ascii')
    # hapus mention, link, hashtag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # remove URL 
    return text.replace("http://", " ").replace("https://", " ")
df['text'] = df.text.astype('str')                
df['text'] = df['text'].apply(remove_text_special)


```

# Menampilkan 5 Data pertama dari dataframe

```python
df.head()
```

# Menghapus elongasi 


```python
#remove kata yang sengaja dipanjangkan
# seperti bisaaaaaaa -> bisa
def spell(word):
    if re.search(r'([a-zA-Z])\1{2,}', word):
        rword = re.sub(r'([a-zA-Z])\1{2,}','\\1', word)
        return rword
    else:
        return word
df['text'] = df['text'].apply(spell)

```

```python
df.head()
```

# menghapus tanda baca

```python
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

df['text'] = df['text'].apply(remove_punctuation)

```

```python
df.head()
```

# menghapus enter tab

```python
#remove whitespace leading & trailing
def remove_whitespace_LT(text):
    return text.strip()

df['text'] = df['text'].apply(remove_whitespace_LT)
```

```python
df.head()
```

# menghapus spasi yang terlalu banyak

```python
def remove_whitespace_multiple(text):
    return re.sub('\s+',' ',text)

df['text'] = df['text'].apply(remove_whitespace_multiple)
```

# menghapus karakter single seperti a b c d

```python
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

df['text'] = df['text'].apply(remove_singl_char)
```

```python
df.head()
```

# Membuat kolom text_token berisi kumpulan kata kata yang dipisah dengan koma

```python
df = df.drop_duplicates()
df = df[['text_real','text','label']]
df['text'] = df['text'].str.lower()

def word_tokenize_wrapper(text):
    return word_tokenize(text)

df['text_tokens'] = df['text'].apply(word_tokenize_wrapper)
```

```python
df.head()
```

# Menghapus stopwords

```python
list_stopwords = stopwords.words('indonesian')
list_stopwords.extend(["yg", "dg", "rt", "dgn", "ny", "d", 'klo', 
                       'kalo', 'amp', 'biar', 'bikin', 'bilang', 
                       'gak', 'ga', 'krn', 'nya', 'nih', 'sih', 
                       'si', 'tau', 'tdk', 'tuh', 'utk', 'ya', 
                       'jd', 'jgn', 'sdh', 'aja', 'n', 't', 
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt',
                       '&amp', 'yah'])

txt_stopword = pd.read_csv("stopwords.txt", names= ["stopwords"], header = None)
list_stopwords.extend(txt_stopword["stopwords"][0].split(' '))
list_stopwords = set(list_stopwords)
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]
df['text_tokens_WSW'] = df['text_tokens'].apply(stopwords_removal) 
```

# Distribusi frekuensi NLTK

```python
def freqDist_wrapper(text):
    return FreqDist(text)

df['text_tokens_fdist'] = df['text_tokens'].apply(freqDist_wrapper)

print('Frequency Tokens : \n') 
print(df['text_tokens_fdist'].head().apply(lambda x : x.most_common()))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

clf =  RandomForestClassifier(random_state=0, max_features ='sqrt',max_depth = 100)
knn = KNeighborsClassifier(n_neighbors=7)
nb = GaussianNB()
dt = DecisionTreeClassifier(max_features ='sqrt',max_depth = 100)
svm = SVC(degree=3)
# clf.fit(X, y)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

normalizad_word = pd.read_csv("dataset_komentar_instagram_cyberbullying.csv")

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

df['text_normalized'] = df['text_tokens_WSW'].apply(normalized_term)


# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in df['text_normalized']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")


# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['text_tokens_stemmed'] = df['text_normalized'].swifter.apply(get_stemmed_term)
```

# Menampilkan kembali gabungan data

```python
def merge_again(text):
    return ' '.join(text)
```

```python
df['done_text'] = df['text_tokens_stemmed'].apply(merge_again)
df
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tfid = TfidfVectorizer(
    min_df = 3,
    max_df = 0.10,
    ngram_range = (1,2)
)
```

```python
count_vect = CountVectorizer(
    min_df = 3,
    max_df = 0.10,
    ngram_range = (1,2)
)
```

```python
tfid_result = tfid.fit_transform(df["done_text"]).toarray()
tfid_df = pd.DataFrame(tfid_result, columns = tfid.get_feature_names())
tfid_df.columns = ["word_" + str(x) for x in tfid_df.columns]
tfid_df.index = df.index
# new_data = pd.concat([, tfid_df], axis=1)
```

```python
cvzr_result = count_vect.fit_transform(df["done_text"]).toarray()
cvzr_df = pd.DataFrame(cvzr_result, columns = count_vect.get_feature_names())
cvzr_df.columns = ["word_" + str(x) for x in cvzr_df.columns]
cvzr_df.index = df.index
# new_data = pd.concat([, tfid_df], axis=1)
```

# Menampilkan Hasil classification

```python
y = df.label
```

```python
X = cvzr_df.to_numpy()
y = df.label
y_pred = cross_val_predict(clf, X, y, cv=5)
conf_mat = confusion_matrix(y_pred, y)
```

```python
print(conf_mat)
print(classification_report(y, y_pred, digits = 4))
```

# Menampilkan hasil knn

```python
print('KNN')
X = cvzr_df.to_numpy()
y = df.label
y_pred = cross_val_predict(knn, X, y, cv=5)
conf_mat = confusion_matrix(y_pred, y)
print(conf_mat)
print(classification_report(y, y_pred, digits = 4))
```

# Menampilkan hasil Decision Tree

```python
print('Decision Tree')
X = cvzr_df.to_numpy()
y = df.label
y_pred = cross_val_predict(dt, X, y, cv=5)
conf_mat = confusion_matrix(y_pred, y)
print(conf_mat)
print(classification_report(y, y_pred, digits = 4))
```
